```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from docx import Document
from docx.shared import Inches

# Define the columns to process
columns = ["Sph-DIFF-R", "Cyl-DIFF-R", "Sph-DIFF-L", "Cyl-DIFF-L", 'Ax-DIFF-R', 'Ax-DIFF-L']
# Create a new Word document
doc = Document()

for column in columns:
    # Step 1: Calculate the population mean
    population_mean = data[column].mean()
    
    # Step 2: Take a sample of 50 with replacement and calculate the sample mean
    sampled_data = data[column].sample(n=50, replace=True, random_state=42)
    sample_mean = sampled_data.mean()
    
    # Step 3: Generate the sample means over 10,000 iterations
    estimate_list = [data[column].sample(n=50, replace=True).mean() for _ in range(10000)]
    estimate_df = pd.DataFrame({'estimate': estimate_list})
    
    # Step 4: Calculate the mean of the sample means
    mean_sample_means = estimate_df['estimate'].mean()
    
    # Step 5: Calculate the standard error
    standard_error = sampled_data.std() / np.sqrt(len(sampled_data))
    
    # Step 6: Plot the histogram and normal curve
    plt.figure(figsize=(10, 6))
    plt.hist(estimate_df['estimate'], bins=25, density=True, alpha=0.4, label="Histogram of sample means")
    
    # Define plot limits and create a range for the normal curve
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, population_mean, standard_error)
    plt.plot(x, p, 'k', linewidth=2, label='Normal curve (CLT)')
    
    # Step 7: Add vertical lines for the population mean, sample mean, and mean of sample means
    plt.axvline(x=population_mean, color='m', linestyle='-', label='Population mean')
    plt.axvline(x=sample_mean, color='r', linestyle='--', label='Sample mean (first sample)')
    plt.axvline(x=mean_sample_means, color='b', linestyle=':', label='Mean of sample means')
    
    # Step 8: Display values on the plot using plt.text
    plt.text(xmin + (xmax - xmin) * 0.02, plt.ylim()[1] * 0.9, f"Population Mean: {population_mean:.4f}", color='m')
    plt.text(xmin + (xmax - xmin) * 0.02, plt.ylim()[1] * 0.85, f"Sample Mean: {sample_mean:.4f}", color='r')
    plt.text(xmin + (xmax - xmin) * 0.02, plt.ylim()[1] * 0.8, f"Mean of Sample Means: {mean_sample_means:.4f}", color='b')
    
    # Step 9: Customize plot labels, title, and legend
    plt.title(f"Sampling Distribution of Sample Mean for {column}")
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    
    # Save the plot as an image file
    image_path = f"C:/forus data/PROJECT 1/{column}_sampling_distribution.png"
    plt.savefig(image_path, bbox_inches='tight')
    plt.close()
    
    # Add the image to the Word document
    doc.add_heading(f"Sampling Distribution of Sample Mean for {column}", level=2)
    doc.add_paragraph(f"Population Mean: {population_mean:.4f}, Sample Mean: {sample_mean:.4f}, Mean of Sample Means: {mean_sample_means:.4f}, Standard Error: {standard_error:.4f}")
    doc.add_picture(image_path, width=Inches(5.5))

# Save the Word document
doc_path = "C:/forus data/PROJECT 1/Sampling_Distribution_Plots.docx"
doc.save(doc_path)

doc_path
```
