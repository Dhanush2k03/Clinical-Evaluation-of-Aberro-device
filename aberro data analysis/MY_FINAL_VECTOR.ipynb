also we check for the outliers for the correspopnding vector calculations using z score

```python
# Define a function to remove outliers based on z-scores
def remove_outliers(df, column, threshold=3):
    # Calculate the z-score for the specified column
    df[f"z_score_{column}"] = stats.zscore(df[column], ddof=1)
    
    # Identify outliers beyond the specified threshold
    outliers = df[(df[f"z_score_{column}"] > threshold) | (df[f"z_score_{column}"] < -threshold)]
    
    # Remove outliers
    df_no_outliers = df.drop(outliers.index)
    
    # Reset the index and drop the z-score column to clean up
    df_no_outliers = df_no_outliers.reset_index(drop=True)
    df_no_outliers = df_no_outliers.drop(columns=[f"z_score_{column}"])
    
    return df_no_outliers

# Remove outliers sequentially for specified columns
df_cleaned_no_outliers = data.copy()
for column in ["SPE-DIFF-R", "J0-DIFF-R", "SPE-DIFF-L", "J0-DIFF-L", "J45-DIFF-R", "J45-DIFF-L"]:
    df_cleaned_no_outliers = remove_outliers(df_cleaned_no_outliers, column)

# Display the final result summary
df_cleaned_no_outliers.describe().T
```

# HYPOTHESIS TESTING

```python
import scipy.stats as stats

# Define the pairs of columns for autorefractor and subjective refraction
column_pairs = [
    ('Spe-ABB-R-VECT', 'Spe-SR-R-VECT'),
    ('J0-ABB-R-VECT', 'J0-SR-R-VECT'),
    ('J45-ABB-R-VECT', 'J45-SR-R-VECT')
]

# Set the significance level
alpha = 0.05

# Iterate through the column pairs
for autorefractor_col, subjective_refraction_col in column_pairs:
    # Extract the relevant columns
    autorefractor_data = data[autorefractor_col]
    subjective_refraction_data = data[subjective_refraction_col]

    # Conduct the paired t-test
    t_statistic, p_value = stats.ttest_rel(autorefractor_data, subjective_refraction_data)

    # Display the results for the current pair
    print(f"Results for {autorefractor_col} vs {subjective_refraction_col}:")
    print(f"T-statistic: {t_statistic}")
    print(f"P-value: {p_value}")
    
    # Interpret the results
    if p_value < alpha:
        print("There is a significant difference between the two sets of readings.\n")
    else:
        print("No significant difference found. The autorefractor readings may be used as a substitute for subjective refraction.\n")
```

# Bland-altman plot

```python
# Import necessary libraries for creating a Word document
from docx import Document
from docx.shared import Inches

# Create a new Word document
doc = Document()

# Define column pairs and titles for the plots
column_pairs = [
    ('Spe-ABB-R-VECT', 'Spe-SR-R-VECT', 'Bland-Altman Plot for SPE M - Right Eye'),
    ('J0-ABB-R-VECT', 'J0-SR-R-VECT', 'Bland-Altman Plot for J0 - Right Eye'),
    ('J45-ABB-R-VECT', 'J45-SR-R-VECT', 'Bland-Altman Plot for J45 - Right Eye'),
    ('Spe-ABB-L-VECT', 'Spe-SR-L-VECT', 'Bland-Altman Plot for SPE M - Left Eye'),
    ('J0-ABB-L-VECT', 'J0-SR-L-VECT', 'Bland-Altman Plot for J0 - Left Eye'),
    ('J45-ABB-L-VECT', 'J45-SR-L-VECT', 'Bland-Altman Plot for J45 - Left Eye')
]

# Loop through each column pair to create Bland-Altman plots and save to Word
for col1, col2, title in column_pairs:
    # Plot the Bland-Altman plot
    mean_values = (data[col1] + data[col2]) / 2
    differences = data[col1] - data[col2]
    mean_diff = np.mean(differences)
    std_diff = np.std(differences)
    loa_upper = mean_diff + 1.96 * std_diff
    loa_lower = mean_diff - 1.96 * std_diff
    
    # Set up the plot with a plain background
    plt.figure(figsize=(10, 6), facecolor='white')
    plt.scatter(mean_values, differences, alpha=0.7, s=50, c='dodgerblue', edgecolor='k', label='Data Points')
    plt.axhline(mean_diff, color='gray', linestyle='--', label=f'Mean Difference: {mean_diff:.2f}')
    plt.axhline(loa_upper, color='red', linestyle='--', label=f'Upper LoA: {loa_upper:.2f}')
    plt.axhline(loa_lower, color='blue', linestyle='--', label=f'Lower LoA: {loa_lower:.2f}')
    plt.title(title)
    plt.xlabel('Mean of Two Measurements')
    plt.ylabel('Difference between Measurements')
    plt.legend()
    plt.grid(False)  # Turn off the grid
    
    # Save the plot as an image
    image_path = f"C:/forus data/PROJECT 1/{title.replace(' ', '_')}.png"
    plt.savefig(image_path, bbox_inches='tight')
    plt.close()
    
    # Add plot title and image to the Word document
    doc.add_heading(title, level=2)
    doc.add_picture(image_path, width=Inches(5.5))

# Save the document
doc_path = "C:/forus data/PROJECT 1/Bland_Altman_Plots.docx"
doc.save(doc_path)

doc_path
```

# Percentage value calculation for vector components

```python


def calculate_percentage_within_2sd(column1, column2):
    # Calculate mean difference and standard deviation
    differences = column1 - column2
    mean_diff = np.mean(differences)
    std_diff = np.std(differences)

    # Compute upper and lower limits of agreement (LoA)
    loa_upper = mean_diff + 1.96 * std_diff
    loa_lower = mean_diff - 1.96 * std_diff

    # Determine LoA range
    loa_range = loa_upper - loa_lower

    # Count data points within LoA range
    within_2sd = np.sum((differences >= loa_lower) & (differences <= loa_upper))

    # Calculate percentage within ±2 SD
    total_data_points = len(column1)
    percentage_within_2sd = (within_2sd / total_data_points) * 100

    return percentage_within_2sd

# Example usage:
percentage_spe_diff_r = calculate_percentage_within_2sd(data['Spe-ABB-R-VECT'], data['Spe-SR-R-VECT'])
print(f"Percentage within ±2 SD for SPE-DIFF-R: {percentage_spe_diff_r:.2f}%")
percentage_spe_diff_r = calculate_percentage_within_2sd(data['J0-ABB-R-VECT'], data['J0-SR-R-VECT'])
print(f"Percentage within ±2 SD for J0-DIFF-R: {percentage_spe_diff_r:.2f}%")
percentage_spe_diff_r = calculate_percentage_within_2sd(data['J45-ABB-R-VECT'], data['J45-SR-R-VECT'])
print(f"Percentage within ±2 SD for J45-ABB-R-VECT: {percentage_spe_diff_r:.2f}%")

percentage_spe_diff_r = calculate_percentage_within_2sd(data['Spe-ABB-L-VECT'], data['Spe-SR-L-VECT'])
print(f"Percentage within ±2 SD for SPE-DIFF-L: {percentage_spe_diff_r:.2f}%")
percentage_spe_diff_r = calculate_percentage_within_2sd(data['J0-ABB-L-VECT'], data['J0-SR-L-VECT'])
print(f"Percentage within ±2 SD for J0-DIFF-L: {percentage_spe_diff_r:.2f}%")
percentage_spe_diff_r = calculate_percentage_within_2sd(data['J45-ABB-L-VECT'], data['J45-SR-L-VECT'])
print(f"Percentage within ±2 SD for J45-ABB-L-VECT: {percentage_spe_diff_r:.2f}%")
```

# Correlation plots and analysis

```python
# Import necessary libraries
from docx import Document
from docx.shared import Inches
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

# Create a new Word document
doc = Document()

# Define pairs of columns and their respective axis labels for the correlation plot
column_pairs = [
    ('Spe-ABB-R-VECT', 'Spe-SR-R-VECT', 'SPE M AHAR', 'SPE M SR', 'Right Eye'),
    ('J0-ABB-R-VECT', 'J0-SR-R-VECT', 'J0 AHAR', 'J0 SR', 'Right Eye'),
    ('J45-ABB-R-VECT', 'J45-SR-R-VECT', 'J45 AHAR', 'J45 SR', 'Right Eye'),
    # You can add right eye columns as needed
]

# Set seaborn style
sns.set(style="white")  # Set style to 'white' to remove background grid

# Loop through column pairs to create correlation plots and add to Word
for autorefractor_col, subjective_refraction_col, xlabel, ylabel, eye_label in column_pairs:
    plt.figure(figsize=(6, 6), facecolor='white')  # Set figure size and background color
    
    # Extract data for the current pair
    x_data = data[autorefractor_col]
    y_data = data[subjective_refraction_col]

    # Calculate Pearson correlation coefficient (r) and r²
    correlation_value, _ = pearsonr(x_data, y_data)
    r_squared = correlation_value ** 2
    
    # Create the scatter plot with regression line and no grid
    sns.regplot(x=x_data, y=y_data, scatter_kws={'s': 50}, line_kws={'color': 'red'})
    plt.grid(False)  # Turn off grid

    # Set custom plot title and labels
    plt.title(f"Correlation between {xlabel} and {ylabel} ({eye_label})")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    # Display r and r² values on the plot
    plt.text(0.05, 0.95, f'r = {correlation_value:.2f}', transform=plt.gca().transAxes, 
             fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))
    plt.text(0.05, 0.88, f'r² = {r_squared:.2f}', transform=plt.gca().transAxes, 
             fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

    # Save the plot as an image
    image_path = f"C:/forus data/PROJECT 1/{autorefractor_col}_vs_{subjective_refraction_col}_correlation.png"
    plt.savefig(image_path, bbox_inches='tight')
    plt.close()
    
    # Add plot title and image to the Word document
    doc.add_heading(f"Correlation between {xlabel} and {ylabel} ({eye_label})", level=2)
    doc.add_picture(image_path, width=Inches(5.5))

# Save the document
doc_path = "C:/forus data/PROJECT 1/Correlation_Plots.docx"
doc.save(doc_path)

doc_path
```
